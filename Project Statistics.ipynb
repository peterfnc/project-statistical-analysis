# USA HATE CRIME ANALYSIS

Our goal is to find out what factors play a role in the levels of hate crime in the United States

## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Import Dataset

#### Downloaded dataset from Kaggle: https://www.kaggle.com/fivethirtyeight/fivethirtyeight-hate-crimes-dataset/downloads/fivethirtyeight-hate-crimes-dataset.zip/108

```python
data = pd.read_csv('hate_crimes.csv')

data.head()
```

## Cleaning & Manipulating

```python
#checking for missing data

data.info()
```

##### In the dataset each row represents one of the US states.  

##### 3 columns have missing data: "share_non_citizen", "hate_crimes_per_100k_splc" and "avg_hatecrimes_per_100k_fbi" <br> as the second 2 mesure the same thing, but the FBI version is an average of multiple years while the Southern Poverty Law Centre only measures a few months I decided to keep the FBI data

```python
data_clean = data.drop(['hate_crimes_per_100k_splc'], axis =1)

data_clean.head()
```

```python
#for the map vizualization had to use the state codes instead of the full name, 
#since they are in order i could just download a csv with the codes

statecodes = pd.read_csv('states.csv')

#Disctrict of Columbia, as an outlier, made my viz harder to read. since its also tiny on the map, i dropped it from the viz 

data_clean['state code'] = statecodes['Abbreviation']

data_formap = data_clean[data_clean['state code'] != 'DC']
```

```python
#for the regression modell we need to take out the rows that contain missing values

data_nonull = data_clean.dropna()

data_nonull.info()
```

### Vizualizing on the Map

```python
import folium


url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
state_geo = f'{url}/us-states.json'


m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=data_formap,
    columns=['state code', 'avg_hatecrimes_per_100k_fbi'],
    key_on='feature.id',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Hate Crime Rate (%)'
).add_to(m)

folium.LayerControl().add_to(m)

m
```

We can see that contrary to the general perception many southern states, <br>
such as Texas, Florida and Georgia are some of the places with the smallest rates of hate crime

At a glance we can't recognize geographical trends, let's try to isolate other factors that might have a role in the levels of hate crime

Based on the Correlation Matrix none of the variables have a particularly strong correlation with the hate crime ratio in a given state. The income inequality (Gini-Index) and the share of Trump supporters seems to have at least some correlation, but not enough to derive conclusions by themselves. Lets further analyse this finding with a multivariate regression modell

```python
Y = data_nonull['avg_hatecrimes_per_100k_fbi']
X = data_nonull[cols[1:-2]]
model = sm.OLS(Y,X)
results = model.fit()
predictions = results.predict(X)
```

```python
print(results.summary())
```

At R2 of 0.804 our modell can be considered fairly good at predicting the hatecrime levels of a state. The OLS results also indicate that the Gini-Index has the strongest explaining power, since its p-value is below 0.05 we can consider it significant
